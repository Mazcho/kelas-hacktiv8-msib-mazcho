Sekarang mari kita hitung rata rata 

```python
mean_2014 = 2.204815e+06  # rata-rata harga properti dpada tahun 2014
std_2014 = 7.174349e+06 # standar deviasi harga properti pada tahun 2014
n = 1230 # jumlah properti di 2014
z = 1.96 # z-score berdasarkan tabel z
se = std_2014 / np.sqrt(n) # menghitung standart eror

# menghitung CI
lower = mean_2014 - z*se
upper = mean_2014 + z*se
print("The lower class boundary is", lower)
print("The upper class boundary is", upper)
```

berdasar hasil diatas, rata rata harga penjualan (sale price) pada tahun 2014 berada di antara 1803868.880130629 sampai 2605761.119869371 dengan peluang atau selang kepercayaan 30,77%

# Hipotesis Testing

Rumusan Masalah :

Dengan mempertimbangkan data NYC Property, apakah tahun 2017  memiliki nilai rata rata yang jual lebih tinggi dari tahun 2002?

H0: Rata-rata harga jual properti di tahun 2017 sama dengan rata-rata harga jual properti di tahun 2002.

H1: Rata-rata harga jual properti di tahun 2017 tidak sama dengan rata-rata harga jual properti di tahun 2002.

```python
yb_17 = df[df["YEAR BUILT"] == 2017]

yb_02 = df[df["YEAR BUILT"] == 2002]

n_17 = len(yb_17)

mean_17 = yb_17["SALE PRICE"].mean()

std_17 = yb_17["SALE PRICE"].std()

(n_17, mean_17, std_17)
```

```python
n_02 = len(yb_02)

mean_02 = yb_02["SALE PRICE"].mean()

std_02 = yb_02["SALE PRICE"].std()

(n_02, mean_02, std_02)
```

```python
sm.stats.ztest(yb_17["SALE PRICE"].dropna(),

           yb_02["SALE PRICE"].dropna(),

          alternative='two-sided')
```

Berdasarkan hasil uji hipotesis dengan menggunakan nilai alpha 0.05, diperoleh nilai p-value lebih besar dari alpha (0.3079 > 0.05) untuk perbandingan harga jual properti antara tahun 2017 dan 2002. Oleh karena itu, hipotesis nol (H0) diterima dan hipotesis alternatif (H1) ditolak. Kesimpulannya adalah tidak terdapat perbedaan yang signifikan dalam harga jual properti antara tahun 2017 dan 2002 di New York City.
