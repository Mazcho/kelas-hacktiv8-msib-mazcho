# Area PLots 

```python
df_can.sort_values(['Total'], ascending=False, axis=0, inplace=True)

df_top5 = df_can.head()

df_top5 = df_top5[years].transpose()

df_top5.head()
```

```python
df_top5.index = df_top5.index.map(int)

df_top5.plot(kind='area',
            stacked=False,
            figsize=(20,10)
            )

plt.title("Immigrations Trend of Top 5 Countries")
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
# transparasi dengan alpha
df_top5.index = df_top5.index.map(int)

df_top5.plot(kind='area',
             alpha=0.25,
            stacked=False,
            figsize=(20,10)
            )

plt.title("Immigrations Trend of Top 5 Countries")
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

# Two Types of Plotting

## Scripting layer 

```python
# transparasi dengan alpha
df_top5.index = df_top5.index.map(int)

df_top5.plot(kind='area',
             alpha=0.25,
            stacked=False,
            figsize=(20,10)
            )

plt.title("Immigrations Trend of Top 5 Countries")
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

## Pake OOP ARTIS LAYER

```python
# transparasi dengan alpha

ax = df_top5.plot(kind='area',
             alpha=0.35,
            figsize=(20,10)
            )

ax.set_title("Immigrations Trend of Top 5 Countries")
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
```

# Pie Chart 

```python
df_continets = df_can.groupby('Continent', axis=0).sum()
print(type(df_can.groupby('Continent',axis=0)))

df_continets.head()
```

```python
df_continets['Total'].plot(kind='pie',
                          figsize=(5,6),
                          autopct='%1.1f%%',
                          startangle=90,
                          shadow = True,
                          )

plt.title('Immigrantion to Canada By Continent [1980-2013]')
plt.axis('equal')

plt.show()
```

```python
colors_list = ['gold','yellowgreen','lightcoral','lightskyblue','lightgreen','pink']
explode_list = [0.1,0,0,0, 0.1, 0.1]

df_continets['Total'].plot(kind='pie',
                          figsize=(15,6),
                          autopct='%1.1f%%',
                          startangle=90,
                          shadow = True,
                           labels = None,
                           pctdistance = 1.12,
                           colors=colors_list,
                           explode=explode_list
                          )

plt.title("immigrant to canada by continent 1980-2013", y=1.12)
plt.axis('equal')

plt.legend(labels=df_continets.index, loc='upper left')
plt.show()
```
