# Confident Intervlas

Rumusan Masalah

Berapa interval dari rata rata jual pada tahun 2014?

```python
c = df.groupby('YEAR BUILT').agg({"SALE PRICE": [np.mean, np.std, np.size]})
c.tail(5)
```

Sekarang mari kita hitung rata rata 

```python
mean_2014 = 2.204815e+06  # rata-rata harga properti dpada tahun 2014
std_2014 = 7.174349e+06 # standar deviasi harga properti pada tahun 2014
n = 1230 # jumlah properti di 2014
z = 1.96 # z-score berdasarkan tabel z
se = std_2014 / np.sqrt(n) # menghitung standart eror

# menghitung CI
lower = mean_2014 - z*se
upper = mean_2014 + z*se
print("The lower class boundary is", lower)
print("The upper class boundary is", upper)
```

berdasar hasil diatas, rata rata harga penjualan (sale price) pada tahun 2014 berada di antara 1803868.880130629 sampai 2605761.119869371 dengan peluang atau selang kepercayaan 69,22%

**Kesimpulan**

Berdasarkan perhitungan diatas, diperoleh hasil sebagai berikut. <br>
Confidence interval = $\mu \pm (z-score \times se)$ 
1.  $\mu - (z_score \times se) = 1803868.880130629 $
2.  $\mu + (z_score \times se) = 2605761.119869371 $
